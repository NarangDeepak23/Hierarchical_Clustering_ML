```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
data=datasets.load_iris()
```

```python
data
```

```python
df=pd.DataFrame(data.data)
```

```python
df
```

```python
df.columns=data.feature_names
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
scaled_df=ss.fit_transform(df)
```

```python
scaled_df
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca_scaled=pca.fit_transform(scaled_df)
```

```python
pca_scaled.shape
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1])
```

# Build a Model for Hierarchichal Clustering
# Create Dendrogram
# Use Agglomerative Clustering


```python
import scipy.cluster.hierarchy as sc
```

```python
# Draw Dendrogram
plt.figure(figsize=(20,10))
plt.title("Dendrogram")

# create dendogram
sc.dendrogram(sc.linkage(pca_scaled,method="ward"))
plt.title("Dendrogram")
plt.xlabel("sample index")
plt.ylabel("Euclidean Distance")
```

```python
# check the vertical longest distance
# LHS is longest one but when horizontal line is drawn it cuts the dendro on RHS
# so RHS ditance is considered as it does not cut any dendro on either side.
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
AC=AgglomerativeClustering(n_clusters=2,affinity="euclidean",linkage="ward")
```

```python
AC.fit(pca_scaled)
```

```python
AC.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=AC.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
clusters=range(2,11)
ss=[]
for i in clusters:
    AC1=AgglomerativeClustering(n_clusters=i,affinity="euclidean",linkage="ward")
    AC1.fit(pca_scaled)
    a=silhouette_score(pca_scaled,AC1.labels_)
    ss.append(a)
ss
```

```python
plt.plot(clusters,ss)
```

```python
# we can see SS is highest at cluster value=2
```
